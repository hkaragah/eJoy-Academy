## How to use python to solve the integral that calculates the volume of a solid
I use a library in Python named sympy to solve the following calculus problem.


### Integral Application: Calculating Volume
Find the volume of a solid that its __base__ is a region enclosed by the parabola $ y=1âˆ’x^2 $ and the x-axis. Crosssections perpendicular to the x-axis are **isosceles triangles** with height equal to the base.

```python
# Importing the required libraries

from sympy import *
```

```python
# Defining the variables that we are going to use for this problem

x, y, b, h, A, V = symbols('x y b h A V')
```

```python
# Defining the relationships between the variables
# See the figure above for further illustration/explanation

y = 1 - x**2
b = y # base of the triangular cross section
h = b # height of the triangular cross section
A = b*h/2 # area of the triangular cross section
```

```python
# Calculating the definite integral
V = integrate(A,(x,-1,1))
print(V)
```
